**Importing Neccessory Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Read Electronics_Sales CSV file**

```python
df=pd.read_csv("electronics_sales.csv")
```

```python
a=df["Product Type"]
b=df["Total Price"]
plt.xlabel("Products")
plt.ylabel("Total Price")
plt.bar(a,b)
plt.show()
```

```python
df.groupby("Product Type")["Total Price"].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
ax = df.groupby("Payment Method")["Total Price"].sum().sort_values(ascending=False).plot(kind='bar',title="Sales by Payment Method",legend=True,ylabel="Total Price",color="green")
```

**Sales** **by** **Product** **Type**

```python
df.groupby("Product Type")["Total Price"].sum().sort_values(ascending=False).plot(kind='bar',
                                                                                  title="Sales by Product Type",
                                                                                  ylabel="Total Sales in Millions",
                                                                                  color="green",
                                                                                  legend=True)
```

**Sales by Payment Method**

```python
df.groupby("Payment Method")["Total Price"].sum().sort_values(ascending=False).plot(kind="line",
                                                                                    title="Sales by Payment Method",
                                                                                    color="green",
                                                                                    ylabel="Total Sales in Millions",
                                                                                    legend=True)
```

```python
df.groupby("Product Type")["Quantity"].sum().sort_values(ascending=False).plot(kind="bar",
                                                                               title="Total Quantity Sold",
                                                                               color="green",
                                                                               ylabel="Quantity",
                                                                               legend=True)
```
